# Living with AIDS

```python
dfl = pd.read_csv("living_with_aids.csv")
```

```python
dfl.info()
```

```python
dfl.head()
```

```python
dfl.rename(columns={"Unnamed: 0": "Country"}, inplace=True)
```

```python
dfl.head()
```

```python
dfl.rename(columns={"Estimated number of people (all ages) living with HIV": "Estimated number of people (all ages) living with HIV 2017"}, inplace=True)
```

```python
dfl.head()
```

```python
dfl = dfl.iloc[1:]
```

```python
dfl.info()
```

```python
dfl = dfl.reset_index()
```

```python
dfl = dfl.drop("index", axis = 1)
```

```python
dfl.head()
```

```python
x = "Estimated number of people (all ages) living with HIV 2017"
dfl = dfl[dfl[x]!="No data"]
```

```python
dfl.head()
```

```python
dfl.info()
```

### Begin plotting

```python
values = []
countries = []
x = "Estimated number of people (all ages) living with HIV 2017"
for country in dfl["Country"]:
    dfx =  dfl[dfl["Country"]==country]
    countries.append(country)
    #converting values to plot appropriate integers
    value = dfx[x].iloc[0]
    r = value.index("[")
    value = value[:r]
    value = value.replace(" ","")
    if "<" in value:
        value = value.replace("<","")
    value = int(value)
    values.append(value)
print(len(countries), len(values))
```

```python
fig = plt.figure(num=None, figsize=(15, 15), dpi=80, facecolor='w', edgecolor='k')
plt.barh(countries, values)
plt.show()
```

```python
print(max(values))
print(countries[values.index(7200000)])
```
