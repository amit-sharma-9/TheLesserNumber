```python
data_geo = data[pd.notnull(data['Location']) ]
```

```python
geo_data=data_geo['Location'].value_counts()
geo_list=data_geo['Location'].value_counts().index
```

```python
geo_list=list(geo_list)
```

```python
for i,value in enumerate(geo_list):
    geo_list[i]=value.strip('()').split(',')
    
    
```

```python
import folium
maryland = folium.Map(location=[39.045753, -76.641273])

for i in range(len(geo_data)):
    lat = geo_list[i][0]
    long = geo_list[i][1]
    radius = geo_data[i] / 45
    
    if geo_data[i] > 500:
        color = "#FF4500"
    else:
        color = "#008080"
    
    popup_text = """Latitude : {}<br>
                Longitude : {}<br>
                Criminal Incidents : {}<br>"""
    popup_text = popup_text.format(lat,
                               long,
                               geo_data[i]
                               )
    folium.CircleMarker(location = [lat, long], popup= popup_text,radius = radius, color = color, fill = True).add_to(maryland)
```

## MAP SHOWS THE LOCATIONS OF CRIME 

```python

import folium
maryland = folium.Map(location=[39.1174, -77.1612],zoom_start=12)
# loop through the 100 crimes and add each to the map
for location, label in zip(geo_list , geo_data):
    lat=location[0]
    lng=location[1]
    radius = label/ 45
    
    if label > 200:
        color = "#FF4500"
    else:
        color = "#008080"
    
    popup_text = """Latitude : {}<br>
                Longitude : {}<br>
                Criminal Incidents : {}<br>"""
    popup_text = popup_text.format(lat,
                               lng,
                               
                               label)
    if label>50:
        folium.CircleMarker(
            [lat, lng],
            radius=radius, # define how big you want the circle markers to be
            color=color,
            fill=False,
            popup=popup_text,

            fill_opacity=0.6
        ).add_to(maryland)
# BIGGER CIRCLE ON MAP REPRESENTS LARGER NO OF ACCIDENTS AT THAT LOCATION 
#ONLY THOSE PLACES ARE PLOT WHICH HAS 50+ RECORDED INCIDENTS
```

```python
maryland.save('index.html')
```

## MAP SHOWS THE  LOCATIONS OF ALL THE CRIME (clustered) 

```python
from folium import plugins
maryland = folium.Map(location=[39.1440, -77.2016],zoom_start=11)
incidents = plugins.MarkerCluster().add_to(maryland)
# loop through the dataframe and add each data point to the mark cluster
for location, label in zip(geo_list , geo_data):
    lat=location[0]
    lng=location[1]
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
        overlay=True
    ).add_to(incidents)
folium.TileLayer('cartodbpositron').add_to(maryland)
```
