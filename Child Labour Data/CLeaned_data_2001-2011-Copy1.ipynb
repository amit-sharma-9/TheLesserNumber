```python
import plotly as pd
import plotly.graph_objs as go
import plotly.plotly as py
import plotly.plotly as tls
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
import pandas as pd
df = pd.read_excel('Census-2001&2011.xlsx')
df
```

```python
df = df.drop(df.columns[4], axis=1)
```

```python
df.columns = ['SNo', 'Name of State/UT' ,'No. of working children in the age group of 5-14 Census 2001', 'No. of working children in the age group of 5-14 Census 2011']
```

```python
df = df.drop(df.index[[0,1,37,38]])
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df = df.drop(df.columns[0],axis=1)
```

```python
df['text'] = df['Name of State/UT'] + '<br>' + df['No. of working children in the age group of 5-14 Census 2001'].astype(str)
```

```python
df
```

```python
data = [
    dict(
        type='choropleth',
        locations=df['Name of State/UT'],
        z=df['No. of working children in the age group of 5-14 Census 2001'],
        colorscale='Portland',
        text = df['text'],
        colorbar = dict(
            title='Thousand Population'
        )
    )
]
```

```python
data
```

```python
layout = dict(
    title='Name of State/UT',
    geo = dict(
        scope='asia', 
        projection = dict(
            type='natural-earth',
     )
   )
)
```

```python
layout
```

```python
flag = dict(data=data, layout=layout)
```

```python
py.iplot( flag, filename='d3-choropleth')
```
